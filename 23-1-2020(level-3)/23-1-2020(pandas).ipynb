#### changing values into null by using loc & iloc

```python
frame=pd.DataFrame(np.random.rand(5,5))
frame
```

```python
frame.iloc[2,2]
```

```python
frame.iloc[2,2]=np.nan
frame
```

```python
frame.iloc[:2,0:]=np.nan
frame
```

```python
frame.isnull()
```

```python
frame.fillna(3)
```

```python
frame.fillna({0:33,1:44,3:55})
```

## plot

```python
help(plt.plot)
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
plt.plot(x,y)
```

```python
plt.plot(x,y,linewidth=2,markersize=10,color='green')
```

```python
plt.plot(x,y,'s',linewidth=2,markersize=10,color='green')
plt.xlabel("x")
plt.ylabel("y")
plt.title("graphing")
plt.show()
```

## bar

```python
a=[1,2,3,4]
b=[10,20,30,40]
plt.bar(a,b,color='pink')
```

```python
plt.bar([1,2,3,4],[10,20,30,40],color="orange")
plt.bar([5,6,7,8],[40,30,20,10],color="red")
```

```python
a1=[2,5,8,3,9]
b1=[8,22,14,5,23]
plt.style.use('ggplot')
plt.plot(a1,b1,color='m')
```

```python
x1=[7,2,4,8,2]
y1=[10,13,6,21,9]
plt.plot(x1,y1,color='orange')
plt.plot(a1,b1,color='k')
```

```python
plt.stackplot(x1,y1,a1)
```

```python
plt.plot(x1,y1,color='orange',label='first plot')
plt.plot(a1,b1,color='k',label='second plot')
plt.legend(loc='lower left')
```

```python
plt.pie(x1,labels=y1)
```

```python
plt.pie(x1,labels=y1,startangle=35)
```

```python
plt.pie(x1,labels=y1,startangle=35,autopct='%1.1f%%')
```
