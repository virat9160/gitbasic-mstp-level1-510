##  creating subplot using numpy with sine

```python
import numpy as np
t=np.arange(0.0,0.5,0.1)
s=np.exp(-t)*np.sin(2*np.pi*t)
plt.subplot(2,2,3)
plt.plot(t,s,color='r',label="graph with sine wave")
plt.xlabel("t")
plt.ylabel("s")
plt.title("graph t vs s")
plt.legend()
plt.show()
```

```python
t=np.arange(0.0,0.5,0.1)
s=np.exp(-t)*np.sin(2*np.pi*t)
plt.subplot(2,2,3)
plt.plot(t,s,color='r',label="first graph with sine wave")
plt.xlabel("t")
plt.ylabel("s")
plt.title("graph t vs s")
plt.legend()
t[0]=np.nan
t[1]=np.nan
plt.subplot(2,1,2)
plt.plot(t,s,label='2nd graph with sine wave')
plt.xlabel('t')
plt.ylabel('s')
plt.title('graph t vs s')
plt.legend()
plt.show()
```

```python
# 3d graph
from mpl_toolkits.mplot3d import axes3d
import matplotlib as style
```

```python
fig=plt.figure()
ax=plt.axes(projection='3d')
plt.show()
```

```python
x=[3,4,2,3]
y=[5,4,99,8]
z=[2,5,4,3]
fig=plt.figure()
ax=fig.add_subplot(212,projection='3d')
ax.plot(x,y,z)
plt.show()
plt.xlabel('x')
plt.ylabel('y')
#plt.zlabel('z')
plt.title('3d projection')
```

```python
x=[3,4,2,3]
y=[5,4,99,8]
z=[2,5,4,3]
dx=np.ones(4)
dy=np.zeros(4)
dz=np.ones(4)
fig=plt.figure()
ax=fig.add_subplot(212,projection='3d')
ax.bar3d(x,y,z,dx,dy,dz,color='m')
plt.show()
plt.xlabel('x')
plt.ylabel('y')
plt.title('3d projection')
```

```python
x=[3,4,2,3]
y=[5,4,99,8]
z=[2,5,4,3]
dx=np.ones(4)
dy=np.zeros(4)
dz=np.ones(4)
fig=plt.figure()
ax=fig.add_subplot(212,projection='3d')
ax.scatter(x,y,z,color='m')
plt.show()
plt.xlabel('x')
plt.ylabel('y')
plt.title('3d projection')
```

```python
# creating subplot using bar graph for a series
import numpy as np
fig,axes=plt.subplots(1,2)
d=pd.Series(np.random.rand(5),index=list("12345"))
print(d)
d.plot.bar(ax=axes[0],color='k')
d.plot.barh(ax=axes[1],color='m',alpha=0.8)
```
