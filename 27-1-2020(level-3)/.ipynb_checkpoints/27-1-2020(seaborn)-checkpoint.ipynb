```python
sb.set(style='whitegrid')
df1=sb.load_dataset('iris')
df1.head()
```

```python
df1['species'].value_counts() 
```

```python
sb.relplot(x='sepal_length',y='petal_length',data=df1)
```

```python
sb.relplot(x='sepal_length',y='petal_length',data=df1,hue='petal_width')
```

```python
sb.relplot(x='sepal_length',y='petal_length',data=df1,hue='petal_width',kind='line')
```

```python
sb.relplot(x='sepal_length',y='petal_length',data=df1,hue='petal_width',col='species')
```

```python
sb.relplot('total_bill',y='tip',data=df,style='smoker',size='size')
```

```python
sb.relplot('total_bill',y='tip',data=df,style='smoker',size='size',col='time')
```

```python
sb.relplot('total_bill',y='tip',data=df,style='smoker',size='size',col='time',legend='full')
```

```python
#linear regression plot
sb.lmplot(x='total_bill',y='tip',data=df)
```

```python
sb.lmplot(x='total_bill',y='tip',data=df,hue='smoker')
```

```python
sb.lmplot(x='total_bill',y='tip',data=df,hue='smoker',col='time')
```
