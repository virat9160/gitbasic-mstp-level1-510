### Bar graph

```python
import matplotlib.pyplot as b
a=list(range(0,51,5))
c=list(range(2,23,2))
b.bar(a,c)
b.xlabel('x-data')
b.ylabel('y-data')
b.title('bar graph')
```

### Scatter plot

```python
import matplotlib.pyplot as s
a=[1,3,5,2]
c=[9,4,6,3]
s.scatter(a,c)
s.xlabel('x-data')
s.ylabel('y-data')
s.title('Scatter plot')
```

### Histogram plot

```python
import matplotlib.pyplot as h
#x=[2,4,6,9]
y=[10,20,30,40]
h.hist(y)
h.xlabel('x-data')
h.ylabel('y-data')
h.title('Histogram plot')
```

```python
#Draw a plot for time and distance by using list comphrension 
#distance=2*t
import matplotlib.pyplot as b
time=[i for i in range(3,6)]
distance=[j*2 for j in time]
b.plot(time,distance,'gs')
b.xlabel('Time-data')
b.ylabel('Distance-data')
b.title('Time & Distance')
```

### matplot ticks
* ticks is values with in a coordinates

```python
import matplotlib.pyplot as pl
x=[9,8,5,6,3,2,1,4,7,10]
y=[5,10,15,20,25,30,35,40,45,50]
pl.plot(y,x,'y')
pl.title('Example of Ticks')
pl.xlabel('Natural numbers')
pl.ylabel("values with multiple by 5")
```

```python
import matplotlib.pyplot as pl
import numpy as np
pl.xticks(np.arange(1,11))
pl.yticks(np.arange(5,51,5))
pl.plot(x,y)
pl.title('Example of Ticks')
pl.xlabel('Natural numbers')
pl.ylabel("values with multiple by 5")
```

#### representing a sing wave
##### amp=sin(wt)
* amp=amplitude
* w=frequency
* t=time

```python
import numpy as np
import matplotlib.pyplot as pl
time=np.arange(1,10,0.1)
amp=np.sin(time)
pl.plot(time,amp)
pl.grid()
pl.title('Sine Wave A=sin(wt)')
pl.xlabel('Time')
pl.ylabel("Amplitude")
```

```python
import pandas as pd
df=pd.read_csv('income.csv')
df
```

```python
import matplotlib.pyplot as pl
income=df.iloc[1,2:]
pl.plot(income)
pl.xlabel('Years')
pl.ylabel("Income")
pl.title('Income of Alaska from 2005-2013')
pl.savefig('Alaska Income')
```

```python
# plot for all states income from 2009-2013
##x-axis=years
##y-axis=income
years=df.columns[-5:]
income=df.loc[:,"2009":]
pl.plot(years,income)
pl.xlabel('Years')
pl.ylabel("Income")
pl.title('Income of all states from 2009-2013')
pl.savefig('All states Income from 2009-2013')
```
