```python
import seaborn as sb
tips=sb.load_dataset('tips')
tips.head()
```

```python
help(sb.factorplot)
```

```python
sb.factorplot(x='total_bill',y='day',data=tips)
```

```python
sb.factorplot(x='total_bill',y='day',data=tips,hue='sex')
```

```python
sb.factorplot(x='total_bill',y='day',data=tips,hue='sex',markers=['x','d'])
```

```python
sb.factorplot(x='total_bill',y='day',data=tips,hue='sex',markers=['x','d'],dodge=True)
```

```python
from numpy import median
sb.factorplot(x='total_bill',y='day',data=tips,hue='sex',markers=['x','d'],dodge=True,
              estimator=median)
```

```python
sb.factorplot('sex','tip',hue='day',data=tips)
```

```python
sb.factorplot('sex','tip',hue='day',data=tips,markers=['x','s','d','o'])
```

```python
help(sb.pointplot)
```

```python
sb.pointplot('total_bill','sex',data=tips)
```

```python
sb.pointplot('total_bill','sex',data=tips,hue='day',markers=['x','s','d','o'])
```

```python
sb.pointplot('total_bill','sex',data=tips,hue='day',markers=['x','s','d','o'],
             order=['Male','Female'])
```

```python
sb.pointplot('total_bill','sex',data=tips,hue='day',color='r')
```

```python
sb.pointplot('total_bill','sex',data=tips,hue='day',palette='PuBuGn')
```

```python
sb.pointplot('total_bill','sex',data=tips,hue='day',markers=['x','s','d','o'],
             linestyles=[':','-.','--',''])
```

```python
sb.pointplot('total_bill','day',data=tips,hue='sex',join=False,dodge=True)
```

```python
sb.pointplot('total_bill','day',data=tips,hue='sex',join=False,dodge=True,estimator=median)
```

```python
import numpy as np
x=np.arange(16).reshape(4,4)
cc=sb.cubehelix_palette()
sb.heatmap(x,cmap=cc)
```

```python
x=np.arange(16).reshape(4,4)
cc=sb.cubehelix_palette(dark=1)
sb.heatmap(x,cmap=cc)
```

```python
x=np.arange(16).reshape(4,4)
cc=sb.cubehelix_palette(dark=1,light=0)
sb.heatmap(x,cmap=cc)
```

```python
x=np.arange(16).reshape(4,4)
cc=sb.cubehelix_palette(dark=1,light=0,as_cmap=True)
sb.heatmap(x,cmap=cc)
```
