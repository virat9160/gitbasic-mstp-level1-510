```python
from sklearn.datasets import load_iris
```

```python
df=load_iris()
```

```python
df
```

```python
df.data
```

```python
df.target_names
```

```python
df.feature_names
```

```python
import pandas as pd
df1=pd.DataFrame(df.data,columns=df.feature_names)
```

```python
df1
```

```python
x=df1.iloc[:,0:4].values
```

```python
x
```

```python
y=df.target
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import VotingClassifier
```

```python
l_clf=LogisticRegression()
r_clf=RandomForestClassifier()
g_NB=GaussianNB()
```

```python
v_clf=VotingClassifier(estimators=[('lr',l_clf),('rf',r_clf),('gb',g_NB)])
```

```python
v_clf.fit(x_train,y_train)
```

```python
#finding accuracy
  #predict(test,prediction)
from sklearn.metrics import accuracy_score
```

```python
for i in (l_clf,r_clf,g_NB,v_clf):
    i.fit(x_train,y_train)
    prediction=i.predict(x_test)
    print(i.__class__.__name__,accuracy_score(y_test,prediction))
```

## Bagging Classifier

```python
d=pd.read_csv('diabetes.csv')
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
x=d.drop('Outcome',axis=1)
x.head()
```

```python
y=d['Outcome']
y.head()
```

```python
b_clf.oob_score_
```

```python
predi=b_clf.predict(x_test)
```

```python
cancer=load_breast_cancer()
```

```python
cancer
```

```python
cancer.data
```

```python
cancer.feature_names
```

```python
x=pd.DataFrame(cancer.data,columns=cancer.feature_names)
```

```python
x.head()
```

```python
encode=LabelEncoder()
binary_y=pd.Series(encode.fit_transform(y))
```

```python
binary_y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,binary_y,test_size=0.25,random_state=1)
```

```python
prediction=ad_clf.predict(x_test)
```

```python
accuracy_score(y_test,prediction)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,prediction)
```
